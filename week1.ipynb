## Draw the graph

```python
import pandas as pd
import networkx as nx
import ast # to convert string to dictionary

G=nx.Graph()
df_nodes=pd.read_csv(r'data\nodes.csv')
columns=df_nodes.iloc[0]
df_edges=pd.read_csv(r'data\edges.csv')
for i in range(len(df_nodes)):
    dictionary = ast.literal_eval(df_nodes.loc[i, ' viz'])
    x = int(dictionary['position']['x'])
    y = int(dictionary['position']['y'])
    colour=dictionary['color']
    size=dictionary['size']
    G.add_node(df_nodes.loc[i, "# index"], pos=(x,y), node_color=colour, node_size=size)
for i in range(len(df_edges)):
    source = df_edges.loc[i, '# source']
    target = df_edges.loc[i, ' target']
    G.add_edge(source, target)

nx.draw(G, nx.get_node_attributes(G, 'pos'),  with_labels=True, )


```
